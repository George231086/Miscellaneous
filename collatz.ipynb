Notebook messing around looking at the Collatz conjecture. Let n0 be a positive integer. We generate the next number in the sequence, n1, by using the rule: if n0 is even we take n1=n0/2, whereas if n0 is odd we take n1=3n0+1. We then repeat this to find n2,n3, etc, finally stopping when we obtain ni=1 for some i. The conjecture is that for any integer n0>0, we will eventually hit 1.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def collatz(n0):
    
    '''
    Basic function to compute collatz sequence. Returns a list with the
    sequence elements.
    '''
    l=[n0]
    while n0!=1:
        if n0%2==0:
            n0*=0.5
        else:
            n0=3*n0+1
        l.append(int(n0))
    return l  
```

Calculate and plot the sequence for n0=31.

```python
l=collatz(31)
print l, "Sequence length: " +str(len(l))
plt.plot(l)
plt.title("Collatz n0=31")
plt.xlabel("ith iteration")
plt.ylabel("ith value")
plt.show()
```

```python
def collatzIterations(n0):
    
    '''
    Basic function to calculate number of iterations. Saves having to store whole
    sequence to calculate stopping time.
    '''
    count=0
    while n0!=1:
        if n0%2==0:
            n0*=0.5 
        else:
            n0=3*n0+1
        count+=1    
    return count
```

Calculate and plot the number of iterations it takes before sequence hits 1, for the first 1000 positive integers.

```python
# vectorize our function so we can pass it numpy arrays.
vCollatzIterations=np.vectorize(collatzIterations)

inputs=np.arange(1,10001)
plot=plt.scatter(inputs,vCollatzIterations(inputs))
plt.title("Collatz iterations")
plt.xlabel("Starting number n0")
plt.ylabel("Iterations")
plt.show()
```

Create some functions and a dataframe to have a mess around with using pandas.

```python
index=np.arange(1,1001)
df=pd.DataFrame(index=index,data=vCollatzIterations(index),columns=["Iterations"])
# Add no column
df['n0']=df.index
# Reorder columns
df=df[['n0','Iterations']]

def isPrime(x):
    
    '''
    Check whether positive integer is prime.
    '''
    
    if x==1:
        return False
    
    i=2
    while i < x:
        if x%i==0:
            return False
        i+=1
    return True

def numberOfPrimes(l):
    
    '''
    Find the number of prime elements in a given list l.
    '''
    
    return len([n for n in l if isPrime(n) and n!=1]) 

df.head()

```

```python
# Add column recording the number of primes in sequence for starting value n0.
df['Number of primes in sequence']=df.index.map(lambda x: numberOfPrimes(collatz(x)))
df.head()
```

```python
# Use boolean mask to create dataframe consisting of prime n0.
mask=df['n0'].map(lambda x: isPrime(x))
primeDf=df[mask]
primeDf.head(5)
```

```python
# Plot a scatter graph for prime n0
plt.scatter(primeDf['n0'],primeDf['Iterations'])
plt.title("Prime collatz iterations")
plt.xlabel("Starting number n0")
plt.ylabel("Iterations")
plt.show()
```

Plot the sequence for n0=871 and 937, the composite and prime n0 $\leq$ 1000 with the highest numbers of iterations.

```python
fig=plt.figure()
ax1=fig.add_subplot(111)

ax1.plot(collatz(871),c='b', marker="s", label='n0=871')
ax1.plot(collatz(937), c='r', marker="o", label='n0=937')
plt.legend(loc='upper left');
plt.title("Collatz n0=871 and 937.")
plt.xlabel("ith iteration")
plt.ylabel("ith value")
plt.show()
```
