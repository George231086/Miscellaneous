```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def collatz(n0):
    
    '''
    Basic function to compute collatz sequence. Returns a list with the
    sequence elements.
    '''
    l=[n0]
    while n0!=1:
        if n0%2==0:
            n0*=0.5
        else:
            n0=3*n0+1
        l.append(int(n0))
    return l  
```

```python
l=collatz(31)
print l, "Sequence length: " +str(len(l))
plt.plot(l)
plt.title("Collatz")
plt.xlabel("ith iteration")
plt.ylabel("ith value")
plt.show()
```

```python
def collatzIterations(n0):
    
    '''
    Basic function to calculate number of iterations. Saves having to store whole
    sequence to calculate stopping time.
    '''
    count=0
    while n0!=1:
        if n0%2==0:
            n0*=0.5
        else:
            n0=3*n0+1
        count+=1    
    return count
```

```python
collatzIterations(31)
```

```python
# vectorize our function so we can pass it arrays.
vCollatzIterations=np.vectorize(collatzIterations)

inputs=np.arange(1,10001)
plot=plt.scatter(inputs,vCollatzIterations(inputs))
plt.title("Collatz iterations")
plt.xlabel("Starting number n0")
plt.ylabel("Iterations")
```

```python
index=np.arange(1,1001)
df=pd.DataFrame(index=index,data=vCollatzIterations(index),columns=["Iterations"])
df['n0']=df.index
df=df[['n0','Iterations']]

def isPrime(x):
    i=2
    while i < x:
        if x%i==0:
            return False
        i+=1
    return True

def numberOfPrimes(l):
    return len([n for n in l if isPrime(n) and n!=1]) 

df.head()

```

```python
df['Number of primes in sequence']=df.index.map(lambda x: numberOfPrimes(collatz(x)))
df.head()
```

```python
primeDf=df[df['n0'].map(lambda x: isPrime(x))]
primeDf.head(5)
```

```python
plt.scatter(primeDf['n0'],primeDf['Iterations'])
plt.title("Prime collatz iterations")
plt.xlabel("Starting number n0")
plt.ylabel("Iterations")
```
